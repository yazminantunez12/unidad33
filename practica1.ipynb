Matplotlib pyplot pyplot. La mayoria de las utilidades de matplotlib se encuentran bajo el pyplot submodulo, y normalmente se importan bajo el plt alias: import matplotlib.pyplot as plt Ahora el paquete pyplot puede denominarse plt. Ejemplo: dibuja una linea en un diagrama desde la posicion (0,0)
hasta la posicion (6,250)

```python
import matplotlib.pyplot as plt
import numpy as np
xpuntos=np.array([0,6])
ypuntos=np.array([0,250])
plt.plot(xpuntos,ypuntos)
plt.show()
```

Matplotlib Plotting. Trazar puntos x-y la funcion plot() se utiliza para dibujar puntos(marcadores) en un diagrama. Por defecto la funcion plot() dibuja la linea de un punto a otro.Esta funcion toma parametros para especificar los puntos en el diagrama.El primer parametro es una matriz que contiene los puntos eje x.El segundo parametro es una matriz que contiene los puntos del eje Y. Si necesitamos trazar una linea de (1,3) a (8,10), tenemos que pasar estos datos en 2 matrices [1,8]y[3,10] a la funcion del trazado plot(). Ejemplo: dibuja una linea en un diagrama desde la posicion (1,3) hasta la posicion (8,10):

```python
import matplotlib.pyplot as plt
import numpy as np
xpuntos=np.array([1,8])
ypuntos=np.array([3,10])
plt.plot(xpuntos,ypuntos)
plt.show()
```
