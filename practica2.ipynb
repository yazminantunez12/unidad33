plotting sin linea
Para trazar solo marcadores, puedes usar notacion de cadena de acceso directo o que significa "anillos"
Ejemplo:Dibuja dos puntos en el diagrama uno en la posicion (1,3) y oto en la posicion (8,10)

```python
import numpy as np
import matplotlib.pyplot as plt
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

Diagrmas de dispercion(scatter)
se emplean para mostrar la relacion entre 2 variables numericas, detectando correlaciones o agrupamientos, Ejemplo:

```python
import numpy as np
import matplotlib.pyplot as plt
altura=([150,160,170,180,190,200])
peso=np.array([50,60,65,80,85,95])
plt.scatter(altura,peso,color="red",marker="x")
plt.ylabel("Altura(cms)")
plt.xlabel("Peso(kg)")
plt.show()
```

Graficos de barras(bar y barh) sirven para comparar categorias o variables tanto en orientacion vertical como horizontal. ejemplo de barras verticales

```python

import matplotlib.pyplot as plt
import numpy as np
categorias=["A", "B", "C", "D"]
valores=[4, 7, 1, 8]

plt.bar(categorias, valores, color="orange")
plt.title("Grafica de barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
```
