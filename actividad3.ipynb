Conceptos básicos de la librería Matplotlib
----------------------------------------------
Matplotlib es una librería de trazados de gráficas de bajo nivel en Python que sirve como un herramienta de visualización. Matplotlib es de código abierto y podemos utilizarlo libremente. Matplotlib está escrito principalmente en Python con algunos segmentos escritos en C.
--------------------------------------------
Instalación de Matplotlib
Si tienes Python y pip ya instalado en tu sistema, instala Matplotlib usando el siguiente comando:
----------------------------------
pip install matplotlib
desde la línea de comando.
-----------------------
Importar Matplotlib

Una vez instalado Matplotlib, impórtalo en tus aplicaciones agregando el módulo import.
--------------------------------
Ejemplo
import matplotlib
------------------
matplotlib.pyplot

-----------------------------------------

Pyplot
La mayoría de las herramientas de Matplotlib se encuentran dentro del submódulo pyplot y normalmente se importan bajo el alias plt:
import matplotlib.pyplot as plt
Ahora el paquete pyplot puede denominarse plt.
